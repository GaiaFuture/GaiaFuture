```python
def subset_only(var):
        # Read in and wrangle user selected variable cluster
    da_v = read_all_simulations(var)
    # feb. ncar time bug
    da = fix_time(da_v)
    # convert xr.ds to xr.da
    da = da[var]
    # weight gridcell dim by global land area
    da_global = weight_landarea_gridcells(da, landarea)
    # weight time dim by days in month
    da_global_ann = yearly_weighted_average(da_global)
    # take global avg for variable over year dimension
    var_avg = da_global_ann.mean(dim='year')
    return var_avg

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# --- Save common variables for future use ---
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# see about appending year into file name to support different avg periods
GPP = subset_only('GPP')
GPP.to_netcdf('GPP.nc')

NBP = subset_only('NBP')
NBP.to_netcdf('NBP.nc') 

TOTVEGC = subset_only('TOTVEGC')
TOTVEGC.to_netcdf('TOTVEGC.nc', index=False)

TLAI = subset_only('TLAI')
TLAI.to_netcdf('TLAI.nc', index=False)

EFLX_LH_TOT = subset_only('EFLX_LH_TOT')
EFLX_LH_TOT.to_netcdf('EFLX_LH_TOT.nc', index=False)

SOILWATER_10CM = subset_only('SOILWATER_10CM')
SOILWATER_10CM.to_netcdf('SOILWATER_10CM.nc', index=False)

QRUNOFF = subset_only('QRUNOFF')
QRUNOFF.to_netcdf('QRUNOFF.nc', index=False)

FSR = subset_only('FSR')
FSR.to_netcdf('FSR.nc', index=False)

FAREA_BURNED = subset_only('FAREA_BURNED')
FAREA_BURNED.to_netcdf('FAREA_BURNED.nc', index=False)

SNOWDP = subset_only('SNOWDP')
SNOWDP.to_netcdf('SNOWDP.nc', index=False)
```

```python
def plot_cluster_panel(param, var_avg):
    '''describe the relationship between the selected variable 
    and parameter(s) between 2005-2010. output a
    cluster plot averaged by year.'''

    # Read in and wrangle user selected parameter cluster
    param_avg  = params[param]

    data = pd.DataFrame({'x': param_avg, 'y': var_avg})

    # Create scatter plot using hvplot
    scatter_plot = data.hvplot.scatter(
        x='x', y='y', color='#62c900ff', alpha=0.8,
        xlabel=param, ylabel="try something", title='2005-2010 Global Average'
    )
    return pn.panel(scatter_plot)

```
