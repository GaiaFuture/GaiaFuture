```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
file='/glade/campaign/cgd/tss/projects/PPE/PPEn11_OAAT/CTL2010/hist/PPEn11_CTL2010_OAAT0000.clm2.h0.2005-02-01-00000.nc'
ds=xr.open_dataset(file)
```

```python
ds.head()
```

```python
##select only the methane
pch4 =  ds['PCH4']

#calculate the average median value
average = pch4.median(dim= 'time')
average
```

```python
plt.scatter(ds.grid1d_lon, ds.grid1d_lat,
            c= average,
            marker = 's',
            plotnonfinite= True)
plt.text(30, 0, 'Sujan-test graphs', fontsize=12, ha='center', va='center', color='black', backgroundcolor='white')


plt.savefig('sujan.png') 
```
